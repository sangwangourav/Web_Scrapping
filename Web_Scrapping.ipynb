```python
import requests

from bs4 import BeautifulSoup
```

```python
url="https://www.imdb.com/chart/top/"
```

```python
res=requests.get(url=url) #fetchimg the data from the url by using get command
```

```python
res #200 means true or yes
```

```python
dir(res) # showing the things present in the url
```

```python
res.content  # giving the html content
```

```python
soup = BeautifulSoup(res.content,'html.parser')# creating the objects and parser means that
# it wil read all the lines from the content is called parsing.
```

```python
soup
```

```python
soup.findAll('td',class_='titleColumn')
```

```python
type(soup.findAll('td',class_='titleColumn')) # result set means list
```

```python
'''[<td class="titleColumn">
       1.
       <a href="/title/tt0111161/" title="Frank Darabont (dir.), Tim Robbins, Morgan Freeman">The Shawshank Redemption</a>
 <span class="secondaryInfo">(1994)</span>
 </td>,
 <td class="titleColumn">'''
# this is a list and its the first row of the list becoz there is a square brackets
```

```python
rows=soup.findAll('td',class_='titleColumn')
```

```python
for row in rows:
    print(row)
```
